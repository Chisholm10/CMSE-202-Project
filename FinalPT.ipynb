```python
import pandas as pd 

# Convert 'Date' column to datetime format
features ['Date'] = pd.to_datetime (features ['Date'])
salesW['Date'] = pd.to_datetime (salesW['Date'])
train_sales['Date'] = pd.to_datetime(train_sales['Date'])

# Filter data for 2010 to 2012
start_date = '2010-01-01'
end_date = '2012-12-31'
features_filtered = features[(features['Date'] >= start_date) & (features['Date'] <= end_date)]
stock_filtered = salesW[(salesW['Date'] >= start_date) & (salesW['Date'] <= end_date)]
weekly_sales = train_sales[(train_sales['Date'] >= start_date) & (train_sales['Date'] <= end_date)]

date_fuel = features_filtered.groupby('Date')['Fuel_Price'].mean().reset_index ()
date_sales = weekly_sales.groupby('Date')['Weekly_Sales'].mean().reset_index ()
merged_data = pd.merge(date_fuel, stock_filtered, on='Date')
merged_data_sales = pd.merge(merged_data, date_sales, on='Date')
merged_data_sales
```

```python
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize= (10, 6))

color = 'tab:green'
ax1.set_xlabel('Date')
ax1.set_ylabel('Fuel Price', color=color)
ax1.plot(merged_data['Date'], merged_data['Fuel_Price'], color=color, label='Fuel Price')
ax1.tick_params (axis='y', labelcolor=color)
ax1.grid()


ax2 = ax1.twinx()

color = 'tab:blue'
ax2.set_ylabel('Opening Price', color=color)
ax2.plot(merged_data['Date'], merged_data['Open'] , color=color, label= 'Opening Price')
ax2.tick_params(axis='y', labelcolor=color)
plt.title("Fuel Price vs Opening Stock Prices from 2010-2012")

```

```python
from sklearn.linear_model import LinearRegression
x1 = merged_data_sales['Fuel_Price'].reshape(-1, 1)
x2 = merged_data_sales['Weekly_Sales'].reshape(-1, 1)
x3 = merged_data_sales['Close'].reshape(-1, 1)


X = np.column_stack((x1, x2))
y = np.column_stack((x3))

# Fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Predictions
y_pred = model.predict(y)

# Plot the data and the regression line
fig, axs = plt.subplots(1, 3, figsize=(15, 5))

for i, (x, label) in enumerate(zip([x1, x2, x3], ['X1', 'X2', 'X3'])):
    axs[i].scatter(x, y, label='Actual Data')
    axs[i].plot(x, y_pred, color='red', label='Linear Regression Line')
    axs[i].set_xlabel(label)
    axs[i].set_ylabel('Y')
    axs[i].legend()

plt.tight_layout()
plt.show()
```
